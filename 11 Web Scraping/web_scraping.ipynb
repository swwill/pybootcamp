```python
res = requests.get("https://en.wikipedia.org/wiki/Grace_hopper")
```

```python
soup = bs4.BeautifulSoup(res.text, "lxml")
```

```python
soup.select(".vector-toc-text span")
```

```python
first_item = soup.select(".vector-toc-text span")[0]
```

```python
first_item
```

```python
for item in soup.select(".vector-toc-text span"):
    print(item.text)
```

```python
for index, line in enumerate(soup.select(".vector-toc-text span")):  # Enumerate provides both index and line
        if index % 2 == 1:  # Check if line number (index + 1) is even
            print(line.text)
```
